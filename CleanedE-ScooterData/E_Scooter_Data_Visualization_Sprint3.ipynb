```python
import pandas as pd
```

```python
df = pd.read_csv(r"E-Scooter_Trips_20240215.csv")
df
```

```python
df=df.drop_duplicates()
df
```

```python
df['Start Community Area Name'] = df['Start Community Area Name'].str.strip()
df['End Community Area Name'] = df['End Community Area Name'].str.strip()
df['Vendor'] = df['Vendor'].str.strip()
df
```

```python
df = df.fillna('')
df
```

```python
df
```

```python
df = df[df.apply(lambda x: x["End Community Area Name"] != '', axis=1)]
```

```python
df
```

```python
df = df[df.apply(lambda x: x["Start Community Area Name"] != '', axis=1)]
df
```

```python
df['Start Centroid Location'] = df['Start Centroid Location'].str.replace('POINT \(', '').str.replace('\)', '').str.replace(' ', ', ')
df
```

```python
df['End Centroid Location'] = df['End Centroid Location'].str.replace('POINT \(', '').str.replace('\)', '').str.replace(' ', ', ')
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.drop('Start Centroid Location', axis=1)
df
```

```python
df = df.drop('End Centroid Location', axis=1)
df
```

```python
# Save the DataFrame to a csv file
df.to_csv('cleaned_e-scooter_data_sprint3.csv', index=False)
```
