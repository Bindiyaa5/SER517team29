**Month with the highest number of rides**

```python
df['month'] = df['Start Time'].dt.month
# Count the occurrences of each month and find the month with the highest number of rides
Most_Preferred_Month = df['month'].value_counts().idxmax()
import calendar
Most_Preferred_Month_Name = calendar.month_name[Most_Preferred_Month]

print(f"The most preferred month for e-scooter riding is: {Most_Preferred_Month_Name}")
```

**Maximum distance travelled in a month**

```python
df.rename(columns={'month': 'Month'}, inplace=True)

# Sum the distances for each month
Monthly_Distance_Sum = df.groupby('Month')['Trip Distance'].sum()

# Find the month with the highest total distance
Most_Distance_Month = Monthly_Distance_Sum.idxmax()

# Translate the numeric month to a month name for readability (optional)
Most_Distance_Month_Name = calendar.month_name[Most_Distance_Month]

print(f"The month with the most distance travelled is: {Most_Distance_Month_Name} with a total distance of {Monthly_Distance_Sum.max()} kilometers.")
```

We got weather information for Chicago from May 2022 to Sep 2023 from the WeatherSpark website. This site gives detailed weather reports. We used it to find out what the weather was like in Chicago during this time. This information will help us look into how the weather might affect E-Scooter usage in the city

```python
# Monthly average temperatures for Chicago from May 2022 to September 2023
monthly_avg_temperatures = [65, 75, 82, 77, 68, 53, 45, 32, 27, 29, 31, 42, 70, 78, 81, 79, 74]

# Extract the month and year from the 'Start Time' column
df['Month'] = df['Start Time'].dt.month
df['Year'] = df['Start Time'].dt.year

# Function to assign temperature based on month and year
def assign_temperature(row):
    if row['Year'] == 2022:
        if row['Month'] < 5:
            return monthly_avg_temperatures[-1]  # Use the last index for months before May 2022
        else:
            index = (row['Month'] - 5) % 17
    elif row['Year'] == 2023:
        index = (row['Month'] + 7) % 17  # Offset by 7 for months in 2023
    return monthly_avg_temperatures[index]

# Apply the function to each row to assign the corresponding weather
df['Weather'] = df.apply(assign_temperature, axis=1)

# Drop the 'Month' and 'Year' columns as they are no longer needed
df.drop(columns=['Month', 'Year'], inplace=True)

# Print the updated DataFrame
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Grouping the e-scooter data by 'Weather' and counting the number of rides ('Trip ID') for each weather
rides_per_weather = df.groupby('Weather')['Trip ID'].count().reset_index()

# Plotting the graph
plt.figure(figsize=(5, 6))
plt.bar(rides_per_weather['Weather'], rides_per_weather['Trip ID'], color='blue')
plt.title('Number of Rides vs. Weather')
plt.xlabel('Weather (Average Temperature °F)')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.show()
```
