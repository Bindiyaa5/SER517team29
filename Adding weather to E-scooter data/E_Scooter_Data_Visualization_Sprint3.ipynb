We got weather information for Chicago from May 2022 to Sep 2023 from the WeatherSpark website. This site gives detailed weather reports. We used it to find out what the weather was like in Chicago during this time. This information will help us look into how the weather might affect E-Scooter usage in the city

```python
# Monthly average temperatures for Chicago from May 2022 to September 2023
monthly_avg_temperatures = [65, 75, 82, 77, 68, 53, 45, 32, 27, 29, 31, 42, 70, 78, 81, 79, 74]

# Extract the month and year from the 'Start Time' column
df['Month'] = df['Start Time'].dt.month
df['Year'] = df['Start Time'].dt.year

# Function to assign temperature based on month and year
def assign_temperature(row):
    if row['Year'] == 2022:
        if row['Month'] < 5:
            return monthly_avg_temperatures[-1]  # Use the last index for months before May 2022
        else:
            index = (row['Month'] - 5) % 17
    elif row['Year'] == 2023:
        index = (row['Month'] + 7) % 17  # Offset by 7 for months in 2023
    return monthly_avg_temperatures[index]

# Apply the function to each row to assign the corresponding weather
df['Weather'] = df.apply(assign_temperature, axis=1)

# Drop the 'Month' and 'Year' columns as they are no longer needed
df.drop(columns=['Month', 'Year'], inplace=True)

# Print the updated DataFrame
df
```

```python
import matplotlib
matplotlib.use('Agg')

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels
from statsmodels.genmod.families import Poisson
from statsmodels.tools.eval_measures import rmse

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

import pandas as pd

# Function to categorize season
def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Winter'

# Function to categorize part of the day
def get_part_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'


# Apply functions to categorize 'started_at' into new columns
df['Day'] = df['Start Time'].dt.day_name()
df['Season'] = df['Start Time'].dt.month.apply(get_season)
df['Time of Day'] = df['Start Time'].dt.hour.apply(get_part_of_day)
df['Same Station Trip'] = (df['Start Community Area Name'] == df['End Community Area Name'])

df = df.groupby(['Vendor', 'Day', 'Season', 'Time of Day', 'Same Station Trip']).agg(
    TMIN_avg=('TMIN', 'mean'),
    TMAX_avg=('TMAX', 'mean'),
    Number_of_Trips = ('TMIN', 'size')
).reset_index()

# Create dummy variables using agg_df
df = pd.concat((df, pd.get_dummies(df['Vendor'])), axis=1)
df = pd.concat((df, pd.get_dummies(df['Day'])), axis=1)
df = pd.concat((df, pd.get_dummies(df['Season'])), axis=1)
df = pd.concat((df, pd.get_dummies(df['Time of Day'])), axis=1)
df = pd.concat((df, pd.get_dummies(df['Same Station Trip'])), axis=1)
df = pd.concat((df, pd.get_dummies(df['TMIN_avg'])), axis=1)
df = pd.concat((df, pd.get_dummies(df['TMAX_avg'])), axis=1)

x1 = pd.DataFrame()
x1 = df
x1.drop(columns=['Vendor','Day', 'Season', 'Time of Day', 'Same Station Trip', 'TMIN_avg', 'TMAX_avg' ], inplace=True)
y = df['Number_of_Trips']
x = df[x1.columns]

# Ensure x is a DataFrame
x = df[x1.columns].copy()

# Check and convert data types of columns
for col in x.columns:
    if x[col].dtypes == 'object':  # Use .dtypes attribute
        x[col] = pd.to_numeric(x[col], errors='coerce')




x = sm.add_constant(x)
pm = sm.GLM(y, x, family=sm.families.Poisson()).fit()
print(pm.summary().as_latex())
print("Poisson regression's RMSE value")
print(np.sqrt(mean_squared_error(y, pm.fittedvalues)))

######################################################################

from sklearn.metrics import mean_squared_error

# Define your train and test sets
size = 0.2
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=size, random_state=42)

# Fit the Poisson regression model on the training data
pm_train = sm.GLM(y_train, x_train, family=sm.families.Poisson()).fit()

# Predict on training data and calculate RMSE
train_preds = pm_train.predict(x_train)
train_rmse = np.sqrt(mean_squared_error(y_train, train_preds))
print(f"Train RMSE: {train_rmse}")

# Predict on test data and calculate RMSE
test_preds = pm_train.predict(x_test)
test_rmse = np.sqrt(mean_squared_error(y_test, test_preds))
print(f"Test RMSE: {test_rmse}")

# Plotting actual vs predicted values for the test set
plt.figure(figsize=(10, 6))
plt.plot(y_test.values[:100], 'b-', label='Actual', linewidth=2)
plt.plot(test_preds.values[:100], 'r--', label='Predicted', linewidth=2)
plt.legend()
plt.title("Actual vs Predicted")
plt.xlabel("Sample Index")
plt.ylabel("Number of Trips")
plt.show()

# Save the plot
plt.savefig("line_comparison.png")
plt.clf()
```
