```python
# Step 1: Load Data
# Assuming the data is in a CSV file named 'ride_data.csv'
data = pd.read_csv('C:/Users/asurapa4/Downloads/merged_rides_with_elevation.csv')
data['started_at'] = pd.to_datetime(data['started_at'])
```

```python
columns_to_delete = ['ride_id','started_at','ended_at','end_station_name','end_station_id','member_casual', 'end_lat','end_lng','Start_Altitude','End_Altitude','Start_Altutude','Elevation_Change''start_station_id','end_altitude','end_Altitude','State_Altitude' ]
data = data.drop(columns=columns_to_delete)
data.head()
```

```python
# Removing rows where rideable_type is 'classic_bike'
df_electric = data[data['rideable_type'] != 'classic']

# Display the DataFrame after removal of 'classic_bike' rows
print("\nDataFrame after removing 'classic_bike' rows:")
print(df_electric)
```

```python
df_electric = df_electric[df_electric['rideable_type'] != 'docked_bike']

# Display the DataFrame after removal of 'docked_bike' rows
print("\nDataFrame after removing 'docked_bike' rows:")
print(df_electric)
```

```python
df_electric.head()
```

```python
duplicates = df_electric.duplicated(subset=['start_lat', 'start_lng'], keep=False)

# Display only duplicates
print("\nDuplicates (if any):")
print(df_electric[duplicates])
```

```python
df_electric_unique = df_electric.drop_duplicates(subset=['start_lat', 'start_lng'])
```

```python
df_electric_unique.to_csv('data_final_electric.csv', index=False)
```

```python
# Sample 1700 items randomly without replacement
sample = df_electric_unique.sample(n=1700, random_state=1)
```

```python
import folium
from folium.plugins import MarkerCluster
# Create a map centered around the average coordinates of the data
average_lat = sample['start_lat'].mean()
average_lng = sample['start_lng'].mean()
map = folium.Map(location=[average_lat, average_lng], zoom_start=11)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(map)

# Add points to the MarkerCluster
for index, row in sample.iterrows():
    folium.Marker(
        location=[row['start_lat'], row['start_lng']],
        popup=row['start_station_name']
    ).add_to(marker_cluster)
```

```python
map.save('electric_map_final.html ')
```
