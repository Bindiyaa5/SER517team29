```python
import pandas as pd

# Load your dataset
data = pd.read_csv('C:/Users/asurapa4/Downloads/merged_rides_with_elevation.csv')
```

```python
data.head()
```

```python
columns_to_delete = ['ride_id','ended_at','end_station_name','end_station_id','member_casual', 'end_lat','end_lng','Start_Altitude','End_Altitude','Start_Altutude','Elevation_Change' ]
data = data.drop(columns=columns_to_delete)
data.head()
```

```python
columns_to_delete = ['started_at']
data = data.drop(columns=columns_to_delete)
data.head()
```

```python
# Removing rows except where rideable_type is 'classic_bike'
df_classic_bike = data[data['rideable_type'] != 'electric_bike']
```

```python

df_classic_bike = data[data['rideable_type'] != 'docked_bike']
print(df_classic_bike)
```

```python
# Removing
df_classic_bike = df_classic_bike.drop_duplicates(subset=['start_lat', 'start_lng'])
print(df_classic_bike)
```

```python
df_classic_bike = df_classic_bike[df_classic_bike['rideable_type'] != 'electric_bike']

# Display the DataFrame after removal of 'classic_bike' rows
print(df_classic_bike)
```

```python
# Sample 1700 items randomly without replacement
sample = df_classic_bike.sample(n=1700, random_state=1)
```

```python
import folium
from folium.plugins import MarkerCluster
# Create a map centered around the average coordinates of the data
average_lat = sample['start_lat'].mean()
average_lng = sample['start_lng'].mean()
map = folium.Map(location=[average_lat, average_lng], zoom_start=11)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(map)

# Add points to the MarkerCluster
for index, row in sample.iterrows():
    folium.Marker(
        location=[row['start_lat'], row['start_lng']],
        popup=row['start_station_name']
    ).add_to(marker_cluster)
```

```python
map.save('classic_map_final.html ')
```
